2.1. Изучите распределение домов от наличия вида на набережную
Постройте график
Сделайте выводы

```python
data = df['waterfront'].value_counts()
names=data.index
values=data.values
data

names=data.index
values=data.values
data

plt.pie(values,autopct='%.1f%%')
plt.title('Распределение домов от наличия вида на набережную')
plt.legend(data.index);

#Менее 1% домов с видом на набережную
```

2.2. Изучите распределение этажей домов

```python
data = df['floors'].value_counts()
names=data.index
values=data.values
data

names=data.index
values=data.values
data

plt.bar(names, values)
plt.title('Распределение домов от кол-ва этажей')
plt.xlabel('floors')
plt.ylabel('Qty');

#Большинство  построено низкоэтажных домов.
```

```python
data= df['floors'].value_counts()
names=data.index
values=data.values
data

plt.figure(figsize=(5,4))
plt.pie(values,autopct='%.1f%%', labels=names)
plt.title('Распределение домов от кол-ва этажей');
#plt.legend(data.index);

#Самые популярные дома с 1 и 2 этажами, вместе составляют 87,5%
```

2.3. Изучите распределение состояния домов


```python
#2.3 Изучите распределение состояния домов
data= df['condition'].value_counts()
names=data.index
values=data.values
data

plt.figure(figsize=(5,4))
plt.pie(values,autopct='%.1f%%')
plt.title('Распределение домов по состоянию')
plt.legend(data.index);

#Наибольшее количетсво домов с состоянием 3 - 64,9%
```

Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями

```python
#Матрица корреляций
import seaborn as sns
corr_matrix=df.corr()

plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix,annot=True,cmap='coolwarm',fmt='.1f');
```

```python

sns.boxplot(x=df['waterfront'], y=df['price'])
plt.title("Взаимосвязь стоимости от наличия набережной")
plt.xlabel('waterfront')
plt.ylable('price')

#Цена выше за дом, при наличии набережной
```

```python
sns.boxplot(x='floors', y='price',
            data=df, width=1, fliersize=2)
plt.title("Взаимосвязь стоимости от этажности")
sns.despine(offset=10, trim=True)

#Самые дорогие, дома с этажностью до 3го
```

```python
sns.boxplot(x=df['bedrooms'], y=df['price'])
plt.title("Взаимосвязь стоимости от спальных комнат")
plt.xlabel('bedrooms')
plt.ylable('price')

#Очень дорогие дома, у которых количество комнат 8
```
